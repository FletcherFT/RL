Taking random actions does not do a very good job at balancing the pole. Let us now apply the Policy Gradient method described above to solve this task!

To start with, our policy will be a rather simple neural network with one hidden layer. We can retrieve the shape of the state space (input) and action space (output) from the environment.

```python
# setup policy network
n_inputs = env.observation_space.shape[0]
n_hidden = 40
n_outputs = env.action_space.n

tf.reset_default_graph()

# States are horizontal position, horizontal velocity, pole angle, and pole angular velocity
states_pl = tf.placeholder(tf.float32, [None, n_inputs], name='states_pl')

actions_pl = tf.placeholder(tf.int32, [None, 2], name='actions_pl')
advantages_pl = tf.placeholder(tf.float32, [None], name='advantages_pl')
learning_rate_pl = tf.placeholder(tf.float32, name='learning_rate_pl')
l_hidden = tf.layers.dense(inputs=states_pl, units=n_hidden, activation=relu, name='l_hidden')
l_out = tf.layers.dense(inputs=l_hidden, units=n_outputs, activation=softmax, name='l_out')

# print network
print('states_pl:', states_pl.get_shape())
print('actions_pl:', actions_pl.get_shape())
print('advantages_pl:', advantages_pl.get_shape())
print('l_hidden:', l_hidden.get_shape())
print('l_out:', l_out.get_shape())
```

```python
# define loss and optimizer (ask about this loss function)
loss_f = -tf.reduce_mean(tf.multiply(tf.log(tf.gather_nd(l_out, actions_pl)), advantages_pl))
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate_pl)
train_f = optimizer.minimize(loss_f)

saver = tf.train.Saver() # we use this later to save the model
```

*Note: As we create our solution, we will make very few assumptions about the cart-pole environment. We aim to develop a general model for solving reinforcement learning problems, and therefore care little about the specific meaning of the inputs and outputs.*

```python
# helper functions

def get_rollout(sess, env, rollout_limit=None, stochastic=False, seed=None):
    """Generate rollout by iteratively evaluating the current policy on the environment."""
    rollout_limit = rollout_limit or env.spec.timestep_limit
    # Initial state of the environment. Changes with seed number. Seed increments with each epoch.
    env.seed(seed)
    s = env.reset()
    states, actions, rewards = [], [], []
    
    for i in range(rollout_limit):
        # Determine action (left or right)
        a = get_action(sess, s, stochastic)
        # Step forward the current environment with action a to get new state s1 and reward r
        # Done happens if the environment failed, so early closing
        s1, r, done, _ = env.step(a)
#         r = r - np.absolute(s1[0])/2.4
        # Append the state, action at state, and reward from executing action at state
        states.append(s)
        actions.append(a)
        rewards.append(r)
        # State becomes new state
        s = s1
        if done: break
    env.seed(None)
    return states, actions, rewards, i+1

def get_action(sess, state, stochastic=False):
    """Choose an action, given a state, with the current policy network."""
    # get action probabilities
    a_prob = sess.run(fetches=l_out, feed_dict={states_pl: np.atleast_2d(state)})
    if stochastic:
        # sample action from distribution
        # The cusum creates an array: [p(going left), p(going left) + p(going right) = 1]
        # rand() picks a number from a uniform distribution between 0 and 1
        # Inequality relation gives either [False True] or [True True].
        # Returns 1 in former case and 0 in latter case with argmax() function.
        return (np.cumsum(np.asarray(a_prob)) > np.random.rand()).argmax()
    else:
        # select action with highest probability
        return a_prob.argmax()

    # advantages help push the network toward favorable simulations and away from unfavorable simulations
def get_advantages(rewards, rollout_limit, discount_factor, eps=1e-12):
    """Compute advantages"""
    returns = get_returns(rewards, rollout_limit, discount_factor)
    # standardize columns of returns to get advantages
    advantages = (returns - np.mean(returns, axis=0)) / (np.std(returns, axis=0) + eps)
    # restore original rollout lengths 
    # i.e. if rollout_limit is 10 and a row in rewards has 5 non zero elements, then advantages should only have 5 non zero
    # elements in that row instead of 10
    advantages = [adv[:len(rewards[i])] for i, adv in enumerate(advantages)]
    return advantages

def get_returns(rewards, rollout_limit, discount_factor):
    """Compute the cumulative discounted rewards, a.k.a. returns."""
    # Size corresponds to number of runs (len(rewards)) and global rollout limit (usually 200 from env.spec.timestep_limit)
    returns = np.zeros((len(rewards), rollout_limit))
    # For each index i and reward r in rewards. 
    # r accounts for all rewards in a single run until failure or final timestep.
    for i, r in enumerate(rewards):
        # len(r)-1 corresponds to the index of the last reward in r
        # we must do this because we know it may be less than rollout_limit
        returns[i, len(r) - 1] = r[-1]
        for j in reversed(range(len(r)-1)):
            returns[i,j] = r[j] + discount_factor * returns[i,j+1]
    return returns
```

```python
# training settings

epochs = 300 # number of training batches
batch_size = 2000 # number of timesteps in a batch
rollout_limit = env.spec.timestep_limit # max rollout length, is 200
discount_factor = 1.00 # reward discount factor (gamma), 1.0 = no discount
learning_rate = 0.003 # you know this by now
early_stop_loss = 0 # stop training if loss < early_stop_loss, 0 or False to disable

# train policy network

try:
    statistics = []
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        print('start training')
        for epoch in range(epochs):
            # generate rollouts until batch_size total timesteps are collected
            states, actions, rewards = [], [], []
            timesteps = 0
            while timesteps < batch_size:
                _rollout_limit = min(rollout_limit, batch_size - timesteps) # limit rollout to match batch_size
                s, a, r, t = get_rollout(sess, env, _rollout_limit, stochastic=True, seed=epoch)            
                states.append(s)
                actions.append(a)
                rewards.append(r)
                timesteps += t
            # compute advantages
            advantages = get_advantages(rewards, rollout_limit, discount_factor)
            # policy gradient update
            loss, _ = sess.run(fetches=[loss_f, train_f], feed_dict={
                states_pl: np.concatenate(states),
                actions_pl: np.column_stack((np.arange(timesteps), np.concatenate(actions))),
                advantages_pl: np.concatenate(advantages),
                learning_rate_pl: learning_rate
            })            
            # validation
            val_rewards = [get_rollout(sess, env, rollout_limit, stochastic=False, seed=(epochs+i))[2] for i in range(10)]
            # store and print training statistics
            mtr = np.mean([np.sum(r) for r in rewards])
            mvr = np.mean([np.sum(r) for r in val_rewards])
            statistics.append((mtr, mvr, loss))
            print('%4d. training reward: %6.2f, validation reward: %6.2f, loss: %7.4f' % (epoch+1, mtr, mvr, loss))
            # early stopping
            if early_stop_loss and loss < early_stop_loss: break
        print('done')
        # save session
        saver.save(sess, 'tmp/model.ckpt')
except KeyboardInterrupt:
    pass    
```

```python
# plot training statistics
statistics = np.array(statistics).T
mean_training_rewards = statistics[0]
mean_validation_rewards = statistics[1]
losses = statistics[2]

plt.figure(figsize=(10,6))
plt.subplot(211)
plt.plot(losses, label='loss')
plt.xlabel('epoch'); plt.ylabel('loss')
plt.xlim((0, len(losses)))
plt.legend(loc=1); plt.grid()
plt.subplot(212)
plt.plot(mean_training_rewards, label='mean training reward')
plt.plot(mean_validation_rewards, label='mean validation reward')
plt.xlabel('epoch'); plt.ylabel('mean reward')
plt.xlim((0, len(mean_validation_rewards)))
plt.legend(loc=4); plt.grid()
plt.tight_layout(); plt.show()
```

### Exercise 3

*In the plot of the training and validation mean reward above, you will sometimes see the validation reward starts out lower than the training reward but later they cross. How can you explain this behavior? [Hint: Do we use the policy network in the same way during training and validation?]*

The policy network is not used in the same way dring training and validation. The stochastic variable is passed as true in get_rollout for training and is passed as false in get_rollout for validation. In the case of validation, the action with the highest probability of occuring is used for the next step of the session. In the case of training, the action is sampled based on its probability of occuring. The implication is that after a few training epochs, the network may be sufficiently trained so  that it has a high probability for selecting the correct action. That is, the network can be relatively certain of what action should take place at each time step. As an example, if at each time step the network has a probability of 80% of selecting an action, then it can be assumed that this is a sufficiently high probability for selecting the correct action. In the case of validation, selecting the actions with high probabilities may lead to simulating the cart-pole balanding problem without any failure. In the case of training, this same network may, in some instances, select the action with a probability of 20%, which may in certain case be the wrong action that leads to the failing the cart-pole balancing problem. Thus training the network leads to a greater number of failurs, and therefore a lesser number of rewards. Validation reward and training reward converge later on as the network becomes better at deciding what the correct action is. This is the case when the network has probabilities of 100% for selecting an action; in this case, regardless of whether the stochastic variable is true or false, the same action will be chosen.

### Exercise 4

*How does the policy gradient method we have used address the exploration-exploitation dilemma (see the previous notebook for definition)?*

The exploration-exploitation dilemma is addressed by sampling actions based on their probabilities. This means that even for a well trained network that has explored its environment, the network may still choose actions that have a low probability of occuring, which can lead to greater future rewards.

http://neuralnetworksanddeeplearning.com/chap1.html#exercise_852508

Write out Equation (22) in component form, and verify that it gives the same result as the rule (4) for computing the output of a sigmoid neuron.

Solution
---------------------
Equation (4) describes the output of a single neuron and is:
$$
\begin{equation}
    a_i = \dfrac{1}{1 + exp(-\sum_{j}w_jx_j -b)}
\end{equation}
$$

Equation (22) describes the activations of a column of neurons and is:

$$
\begin{equation}
    a' = \sigma(wa + b)
\end{equation}
$$

where $w$ is a n-by-m matrix, $a$ is a m-by-1 vector, $b$ is a n-by-1 vector, and $a'$ is a n-by-1 vector.

To write equation (22) in component form, we first instert the definition for the sigmoid function and obtain

$$
\begin{align}
    a' &= \dfrac{1}{1 + exp(-(wa + b))} \\
     &= \dfrac{1}{1 + exp(-wa - b)}
\end{align}
$$

Since $w$ is a n-by-m matrix, $a$ is a m-by-1 vector, $b$ is a n-by-1 vector, and $a'$ is a n-by-1 vector, we can rewrite what we obtain in component form:

$$
\begin{equation}
    a' = 
    \begin{bmatrix}
    a_1 \\ a_2 \\ \vdots \\ a_{n-1} \\ a_{n}
    \end{bmatrix}
    = 
    \begin{bmatrix}
    \dfrac{1}{1 + exp(-\sum_{j = 1}^{m}w_{1,j}x_j -b_1)} \\
    \dfrac{1}{1 + exp(-\sum_{j = 1}^{m}w_{2,j}x_j -b_2)} \\
    \vdots \\
    \dfrac{1}{1 + exp(-\sum_{j = 1}^{m}w_{n-1,j}x_j -b_{n-1})} \\
    \dfrac{1}{1 + exp(-\sum_{j = 1}^{m}w_{n,j}x_j -b_{n})}
    \end{bmatrix}
\end{equation}
$$
